```
# Importing required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```
# Reading the csv file
df = pd.read_csv('https://raw.githubusercontent.com/midhunmohank/INFO6105/main/advertising.csv')

```

```
# Checking for missing values in the dataset
print(df.isnull().sum())

```

```
# Displaying the first few rows of the dataset
print(df.head())

```

```
# Displaying the summary statistics of the numerical columns in the dataset
print(df.describe())

```

```
# Visualizing the distribution of the numerical columns using histograms
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
sns.histplot(df['TV'], bins=10)
plt.title('Distribution of TV')

plt.subplot(2,2,2)
sns.histplot(df['Radio'], bins=10)
plt.title('Distribution of Radio')

plt.subplot(2,2,3)
sns.histplot(df['Newspaper'], bins=10)
plt.title('Distribution of Newspaper')

plt.subplot(2,2,4)
sns.histplot(df['Sales'], bins=10)
plt.title('Distribution of Sales')

plt.show()

```

```
# Visualizing the correlation between the numerical columns
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

```

```
# Visualizing the relationship between each numerical column and the target variable using pairplot
plt.figure(figsize=(10,8))
sns.pairplot(df, x_vars=['TV', 'Radio', 'Newspaper'], y_vars='Sales', height=4, aspect=1, kind='scatter')
plt.show() 

```
